```python
import timeit
CNN_cell=tf.keras.layers.Conv2D(filters=100,kernel_size=2,strides=(1,1))
@tf.function
def CNN_fn(image):
  return CNN_cell(image)
image=tf([100,200,200,3])
CNN_cell(image)
CNN_fn(image)
print("time required for performing the CNN",timeit.timeit(lambda:CNN_cell(image),number=10))
print("time required CNN graph mode:",timeit.timeit(lambda:CNN_fn(image),number=10))
```
